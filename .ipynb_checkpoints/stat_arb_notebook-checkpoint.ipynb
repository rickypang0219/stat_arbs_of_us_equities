```python
import numpy as np 
import pandas as pd 
import yfinance as yf 
import matplotlib.pyplot as plt
```

```python
# Download the data for JPM, XLF, and SPY
jpm = yf.download("JPM", start="2017-01-01", end="2023-07-07")
xlf = yf.download("XLF", start="2017-01-01", end="2023-07-07")
spy = yf.download("SPY", start="2017-01-01", end="2023-07-07")
```

```python
jpm.tail()
```

```python
plt.figure(figsize=(20,5))
plt.title("Stock Prices")
plt.plot(jpm.index, jpm['Close'], label='JPM')
plt.plot(xlf.index, xlf['Close'], label="XLF")
plt.plot(spy.index, spy['Close'], label="SPY")
plt.legend()
```

```python
def clean(df, ticker):
    # Add a new 'Ticker' column with the given ticker symbol
    df['Ticker'] = ticker
    # Create the 'NextOpen' column by shifting the 'o' column by -1
    df['NextOpen'] = df['Open'].shift(-1)
    # Select the desired columns
    df = df[['Date', 'Close', 'Open', 'Ticker', 'NextOpen']]
    return df
```

```python
JPM_clean =  clean(jpm, "JPM")
XLF_clean =  clean(xlf, "XLF")
SPY_clean =  clean(spy, "SPY")
```

```python
concat_df = pd.concat([JPM_clean, XLF_clean, SPY_clean], axis=0).sort_index()
concat_df
```

```python
def calculate_returns(group):
    group['Return'] = np.concatenate([[np.nan], np.diff(np.log(group['Close']))])
    group['ReturnO'] = np.concatenate([[np.nan], np.diff(np.log(group['Open']))])
    group['ReturnTC'] = np.concatenate([[np.nan], np.diff(np.log(group['NextOpen']))])
    return group
```

```python
allPrices = concat_df.groupby('Ticker').apply(calculate_returns).reset_index(drop=True)
```

```python
modelData = allPrices.pivot(index='Date', columns='Ticker', values='Return')
modelData.reset_index(inplace=True)
modelData = modelData.iloc[1:, :]
modelData.tail()
```

```python
plt.figure(figsize=(20,5))
plt.title("Return of Stocks")
plt.plot(modelData['Date'], modelData['JPM'].cumsum(), label='JPM')
plt.plot(modelData['Date'], modelData['XLF'].cumsum(), label="XLF ETF")
plt.plot(modelData['Date'],modelData['SPY'].cumsum(), label="SPY ETF")
plt.legend()
```

# Marco Regression
In this part, we will build a factor model to explain the returns of JPM in terms of SPY ETF and XLF ETF. This model analyse the how does the JPM returns relate to systematic movement of market, governed by XLF and SPY ETF. After removing the systematic movement of JPM returns, the residual is what factor model cannot explain, which is the idosyncratic term of the returns series itself. In the following, we have to use Orsteain Uhlenbeck process to model the residual of JPM returns.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X  = modelData[['XLF',"SPY"]]
X  = X.to_numpy()
y = modelData['JPM'].to_numpy()
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# compute predicted y from Linear regression omdel 
y_pred = model.predict(X)
```

```python
# params and intercept
print(model.coef_, model.intercept_)
```

```python
# residual from linear regression y_pred and true y 
residual = y - y_pred

plt.figure(figsize=(20,5))
plt.title('Residual of Linear Regression Model')
plt.plot(modelData['Date'],residual, "o", label='Residual')
plt.legend()

plt.figure(figsize=(20,5))
plt.title('Cumsum of JPM;s Residual')
plt.plot(modelData['Date'],np.cumsum(residual), label='Residual')
plt.legend()
```

# Orstein Uhlenbeck Process

An OU process is an AR(1) model that has mean reversion behaviour, meaning that the process fluctuates around its mean with a period. The OU process can be modedl by a stochastic process 

$$
dX_t = \kappa ( \mu - X_t) dt + \sigma dW_t
$$
where $W_t$ is Brownian motion, $\mu$ is the equilibrium mean, and $\kappa$ indicates the fast a period of the mean reversion process. We can numerically intergrate the $X_t$ by following transformation 

$$
\begin{align}
dX_t + \kappa X_t dt &= \kappa \mu dT + \sigma dW_t \\ 
d(e^{\kappa t}X_t) &= e^{\kappa t} \kappa \mu dt + e^{\kappa t} \sigma dW_t \\ 
e^{\kappa t}X_t  -  X_0 & =  \int_0^t e^{\kappa s} \kappa \mu ds + \int_0^t e^{\kappa s} \sigma dW_s \\ 
X_t   & = X_0e^{-\kappa t}  +   \mu (1 - e^{-\kappa t} ) + \sigma \int_0^t e^{-\kappa (s-t)}  dW_t \\ 
\end{align}
$$


We can also compare the expressions of OU process and AR(1) process
$$
\begin{align}
dX_t = X_{t+ dt} - X_t &= -\kappa dt X_t +  ( \mu \kappa dt + \sigma dW_t) \\
X_{t + dt}  &= \phi X_t + \epsilon_t
\end{align}
$$
we can get 
$$
\phi = (1- \kappa dt), ~ \epsilon_t =  \mu \kappa dt + \sigma dW_t
$$

Because the data points are daily data and there are 252 trading days of a year, the infintesimal time difference $dt = 1/252$. If we apply AR(1) to regress the cumulative sum of residual, the constant of the first lag $\phi$ relates to the period of mean reversion $\kappa$. After using Linear regression to perform AR(1), we can compute the mean of the cumulative sum of residuals. The AR(1) reads as follows

$$
X_t = \beta_0 + \beta_1 X_{t-1} + \epsilon_t
$$
Compare the linear regression model to OU process, we can obtain $\beta_1 = (1-\kappa dt) $ and $\beta_0 = \mu \kappa dt $. Hence, we can rewrite the mean $\mu$ in terms of $\beta_0$ and $\beta_1$ 

$$
\beta_1 = (1- \beta_0 / \mu) \Rightarrow \mu = \frac{\beta_0}{1- \beta_1}
$$

```python
ar_1_model = LinearRegression()
y = np.cumsum(residual)[1:]
X = np.cumsum(residual)[:-1]
dt = 1/252
```

```python
ar_1_model.fit(X.reshape(-1,1),y)
```

```python
ar_1_model.coef_, ar_1_model.intercept_
```

```python
kappa = (1 - ar_1_model.coef_)/ dt
mu = (ar_1_model.intercept_) / (1 - ar_1_model.coef_)
```

```python
kappa, mu
```

```python
var_eps = np.var( y - ar_1_model.predict(X.reshape(-1,1)), ddof=1) / (1 - ar_1_model.coef_**2)
var_eps
```

```python
sigma = np.sqrt(2* kappa * var_eps)
sigma
```

```python
sigma_eq = np.sqrt(var_eps)
sigma_eq
```

```python
s_score = (X - mu) / sigma_eq
```

```python
plt.figure(figsize=(20,5))
plt.title("S score time series")
plt.plot(modelData.Date[:-1], s_score,label='s score')
plt.axhline(y=-1.25, color='r', linestyle='-', label='Buy to Open')
plt.axhline(y=-0.75, color = "r", linestyle='-.', label='Close Long Position')

plt.axhline(y=1.25,color='purple', linestyle='-', label='Short to Open')
plt.axhline(y=0.5,color='purple' ,linestyle='-.', label='Close Short Position')
plt.legend()
```
